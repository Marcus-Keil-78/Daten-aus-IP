```python
## Hilfsfunktionen
# Sprachkürzel suchen
def get_selected_values(combo_widget_1, combo_widget_2):
    '''
    Gibt die Values anhand der Keys zurück
    '''
    selected_sprache_1 = combo_widget_1.get()
    selected_sprache_2 = combo_widget_2.get()
    
    value_1 = sprachen_dict.get(selected_sprache_1)
    value_2 = sprachen_dict.get(selected_sprache_2)
     
    return value_1, value_2

def standartwerte():
    global abfrage_daten_holen
    ziel_sprache_combo.set("Deutsch")
    quelle_sprache_combo.set("Englisch")
    RadioButtton_VAR.set("IPv4")
    Datum_ausgabe.configure(text="")
    IP_Adresse_LAN_ausgabe.configure(text="")
    IP_Adresse_WAN_ausgabe.configure(text="")
    Stadt_ausgabe.configure(text="")
    Postleitzahl_ausgabe.configure(text="")
    Region_ausgabe.configure(text="")
    Land_ausgabe.configure(text="")
    Kontinent_ausgabe.configure(text="")
    Provider_ausgabe.configure(text="")
    Längengrad_ausgabe.configure(text="")
    Breitengrad_ausgabe.configure(text="")
    Währung_ausgabe.configure(text="")
    Währung_Symbol_ausgabe.configure(text="")
    zeitzone_ausgabe.configure(text="")
    abfrage_daten_holen=False
standartwerte()
```

```python
# Sprachen für die Überstzung
def uebersetze_text(text):
    #holen der daten zum Überstzen
    sprache_quelle, sprache_ziel = get_selected_values(quelle_sprache_combo, ziel_sprache_combo)
    # Stelle sicher, dass du eine API-Schlüssel für die Übersetzungsdienste hast
    # Ersetze 'DEIN_API_SCHLUESSEL' durch deinen tatsächlichen API-Schlüssel
    api_schluessel = 'DEIN_API_SCHLUESSEL'
    
    base_url = 'https://translate.googleapis.com/translate_a/single'
    
    params = {
        'client': 'gtx',
        'sl': sprache_quelle,
        'tl': sprache_ziel,
        'dt': 't',
        'q': text,
    }

    try:
        # Führe die Übersetzung durch
        try:
            antwort = requests.get(base_url, params=params).json()
        except:
            messagebox.showerror("Webdienst", f'Es gab ein Fehler beim aufrufen der URL "{base_url}"\n\n{e}')
            return None
        
        # Extrahiere den übersetzten Text aus der Antwort
        uebersetzt_text = antwort[0][0][0]

        return uebersetzt_text

    except Exception as e:
        messagebox.showerror(f"Fehler bei der Übersetzung:\n\n{e}")
        return None
```

```python
def get_wan_ips():
    '''
    Gibt die IPv4 und IPv6 Adresse des Providers zurück:
    Zuerst wird die IPv4 dann die IPv6 zurück gegeben
    '''
    
    try:
        response_ipv4 = requests.get('https://api.ipify.org?format=json')
        data_ipv4 = response_ipv4.json()
        wan_ip_ipv4 = data_ipv4.get('ip')
    except:
        wan_ip_ipv4 = None

    try:
        response_ipv6 = requests.get('https://api6.ipify.org?format=json')
        data_ipv6 = response_ipv6.json()
        wan_ip_ipv6 = data_ipv6.get('ip')
    except:
        wan_ip_ipv6 = None

    return {"IPv4": wan_ip_ipv4,
            "IPv6": wan_ip_ipv6}
```

```python
def ausgabe():
    global abfrage_daten_holen, daten
    variante_IP_Adresse=RadioButtton_VAR.get()
    
    if get_wan_ips()[RadioButtton_VAR.get()] == None:
        messagebox.showerror('Funktion "ausgabe"', f'Keine {variante_IP_Adresse}-IP-Adresse vorhanden')
        abfrage_daten_holen=False
    else:
        try:
            eigene_ip=socket.gethostbyname(socket.gethostname())
        except Exception as e:
            messagebox.showerror("Fehler",f'Es trat ein Fehler beim lesen der "LAN-IP" auf.\n\n{e}')
        
        # anmelden an ipinfo
        try:
            ip_daten=ipinfo.getHandler(token_daten)
        except Exception as e:
            messagebox.showerror("Webdienst",f'Es trat ein Fehler mit dem "ipinfo.getHandler()" auf\n\n{e}')
        # Daten anhand der IP holen
        try:
            daten=ip_daten.getDetails(get_wan_ips()[variante_IP_Adresse])
        except Exception as e:
            messagebox.showerror("Webdienst",f'Es trat ein Fehler mit dem "ipinfo.getDetails()" auf\n\n{e}')
        
        # Konfigurieren der ausgabe
        Datum_ausgabe.configure(text=datetime.now().strftime("%d.%m.%Y"))
        IP_Adresse_LAN_ausgabe.configure(text=eigene_ip)
        IP_Adresse_WAN_ausgabe.configure(text=daten.ip)
        Stadt_ausgabe.configure(text=uebersetze_text(daten.city))
        Postleitzahl_ausgabe.configure(text=daten.postal)
        Region_ausgabe.configure(text=uebersetze_text(daten.region))
        Land_ausgabe.configure(text=uebersetze_text(daten.country_name))
        Kontinent_ausgabe.configure(text=uebersetze_text(daten.continent["name"]))
        zeitzone_ausgabe.configure(text=daten.timezone)
        Provider_ausgabe.configure(text=daten.org)
        Längengrad_ausgabe.configure(text=daten.longitude)
        Breitengrad_ausgabe.configure(text=daten.latitude)
        Währung_ausgabe.configure(text=daten.country_currency["code"])
        Währung_Symbol_ausgabe.configure(text=daten.country_currency["symbol"])
        
        abfrage_daten_holen=True
```

```python
def speichern():
    if abfrage_daten_holen:    
        path=filedialog.asksaveasfilename(initialfile="Meine Daten nach der IP-Adresse", defaultextension=".txt", filetypes=[("Textdatei", ".txt")], initialdir=f'C:/Users/{os.getlogin()}/Desktop')
        if path != "":
            message=f'Datum: {Datum_ausgabe.cget("text")}\nIP-Adresse LAN: {IP_Adresse_LAN_ausgabe.cget("text")}\nIP-Adresse WAN: {IP_Adresse_WAN_ausgabe.cget("text")}\nStadt: {Stadt_ausgabe.cget("text")}\nRegion: {Region_ausgabe.cget("text")}\nLand: {Land_ausgabe.cget("text")}\nKontinent: {Kontinent_ausgabe.cget("text")}\nProvider: {Provider_ausgabe.cget("text")}\nLängengrad: {Längengrad_ausgabe.cget("text")}\nBreitengrad: {Breitengrad_ausgabe.cget("text")}\nWährung: {Währung_ausgabe.cget("text")}\nWärung Symbol: {Währung_Symbol_ausgabe.cget("text")}'
            with open(path, "w") as file:
                file.write(message)
    else:
        messagebox.showinfo("Daten Speichern", "Keine Daten zum Speichern vorhanden.")
```
